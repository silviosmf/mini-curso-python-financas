## Princpios de Programao e po Python

### Programa

Conjunto de instrues que contm as operaes necessrias para, a partir de dados inseridos, obter um resultado que ser disponibilizado por algum dispositivo de sada (XAVIER, 2018).

De modo geral, os programas de computadores so executados a partir da recepo de informaes externada (Entrada), da execuo das informaes, com as quais foram previamente defindas pelo programados (Processamento) e do retorno dos dados aps a finalizao das instrues (Sada)



#### Entrada
* Dados envidados por teclado, leitores pticos, leitura de arquivos texto ou de imagens...

#### Processamento
* Conjunto de instrues executados por uma linguagem de programao que foram anteriorementes escritas por programados.

#### Sada
* Retorno visuais nos monitores dos computadores, impresses, arquivos gravados em disco...

### Linguagem de Programao

Conjunto de repsentaes e/ou regras utilizadas para criar programas. Por meio delas voc estabelece uma comunicao com o computador, fazendo com que ele compreenda e execute o que voc determinar (XAVIER, 2018).

* #### Produtividade do desenvolvedor

Python aumenta a produtividade do desenvolvedor comparado a linguagens como C, C++ e Java. O cdigo em pythonn , geralmente, 1/3 a 1/5 do tamanho do cdigo equivalente s linguagens destacadas.

* #### Curva de aprendizagem

Python  uma linguagem de sintaxe simples comparados  outras como C, C++, Java e R.

* #### Bibliotecas

Python possui uma infinidade de funcionalidades pr-compiladas e portveis, conhecidas como bibliotecas padro.

* #### Anlise e Cincia da Dados

Python  amplamente utilizada para a anlise e cincia de dados por possuir bibliotecas consolidadas com Numpy, Pandas, Scikit-learn e Statsmodels.

#### Instalao do intepretador Python

Como Python  uma linguagem intepretada, e, como tal,  um tipo de programa que executa outros programas. Quando voc escreve programas em Python, o interpretador l seu programa e executa as instrues.

Portanto, para comearmos a escrever nossos prprios programas devemos instalar no interpretador que est disponvel gratuitamente no site: https://www.python.org/. Porm, para facilitar a instalao do interpretador Python, bem como demais ferramentas que utilizamos no curso, sugere-se a instalao do Software de Gerenciamento de Pacotes Anaconda. site: https://www.anaconda.com/products/individual

O Projeto Jupyter  uma organizao sem fins lucrativos criada para "desenvolver software de cdigo aberto, padres abertos e servios para computao interativa em dezenas de linguagens de programao". Originado do IPython em 2014, o Projeto Jupyter suporta ambientes de execuo em dezenas de linguagens de programao. O nome do projeto  uma referncia s trs principais linguagens de programao suportadas por Jupyter, Julia, Python e R.
