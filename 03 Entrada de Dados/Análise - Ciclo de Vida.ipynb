```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
    caminho_arquivos = '../dados/cvm'
    # Carregar DFC
    dfc = pd.read_csv(f'{caminho_arquivos}/DFC_MI.csv', sep=';', encoding='ISO-8859-1', decimal=',')
```

```python
list_op_invest_fin = []

for ano in range(2015,2021):
    for mes in range(3,15,3):
        list_op_invest_fin.append(dfc[(dfc['CNPJ_CIA'] == '47.960.950/0001-21') & 
                    (pd.to_datetime(dfc['DT_REFER']).dt.month == mes) &
                    (pd.to_datetime(dfc['DT_REFER']).dt.year == ano) &
                    ((dfc['CD_CONTA'] == '6.01') |
                    (dfc['CD_CONTA'] == '6.02') |
                    (dfc['CD_CONTA'] == '6.03'))])
```

```python
len(list_op_invest_fin)
```

```python
list_fc = []

for op_invest_fin in list_op_invest_fin:
    fc = dict()
    fc['data'] = op_invest_fin.iloc[0]['DT_REFER']
    fc['operacional'] = op_invest_fin.iloc[0]['VL_CONTA']
    fc['investimento'] = op_invest_fin.iloc[1]['VL_CONTA']
    fc['financiamento'] = op_invest_fin.iloc[2]['VL_CONTA']
    list_fc.append(fc)
```

```python
df_fc = pd.DataFrame(list_fc)
```
