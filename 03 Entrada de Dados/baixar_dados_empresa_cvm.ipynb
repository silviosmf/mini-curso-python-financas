# Baixar e criar planilhas agrupadas para cada demonstração

```python
import pandas as pd
import wget
from zipfile import ZipFile
import os
```

URL onde estão as demonstrações contábeis da B3

```python
url_itr = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/'
url_dfp = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/'
```

Definir período para Download

```python
inicio_coleta = 2011
fim_coleta = 2021
```

Definir caminho para salvar arquivos

```python
caminho_salvar_arquivos = 'dados/cvm'
```

Definir as lista de arquivos que serão baixados

```python
arquivos_itr = []
arquivos_dfp = []
for ano in range(inicio_coleta,fim_coleta+1):
    arquivos_itr.append(f'itr_cia_aberta_{ano}.zip')
    arquivos_dfp.append(f'dfp_cia_aberta_{ano}.zip')
```

Realizar download dos arquivos ITR (ZIPs)

```python
for arquivo in arquivos_itr:
    wget.download(url_itr+arquivo)
```

Realizar download dos arquivos DFP (ZIPs)

```python
for arquivo in arquivos_dfp:
    wget.download(url_dfp+arquivo)
```

Extrair arquivos

```python
# for arquivo in arquivos_itr:
#     ZipFile(arquivo,'r').extractall(caminho_salvar_arquivos)
    
for arquivo in arquivos_dfp:
    ZipFile(arquivo,'r').extractall(caminho_salvar_arquivos)
```

Remover Arquivos (ZIP)

```python
# for arquivo in arquivos_itr:
#     try:
#         os.remove(arquivo)
#         print(f'O {arquivo} foi removido com sucesso')
#     except:
#         print('Arquivo não encontrado')
        
for arquivo in arquivos_dfp:
    try:
        os.remove(arquivo)
        print(f'O {arquivo} foi removido com sucesso')
    except:
        print('Arquivo não encontrado')
```

Agrupar das demonstrações anuais em um único arquivo por demonstração

```python
demonstrativos = ['BPA','BPP','DRE','DFC_MD','DFC_MI','DMPL','DRA','DVA']

for nome_demo in demonstrativos:
    # Demonstrativo
    demonstrativo = pd.DataFrame()
    
    for ano in range(inicio_coleta, fim_coleta+1):
        demo_itr = pd.read_csv(f'{caminho_salvar_arquivos}/itr_cia_aberta_{nome_demo}_con_{ano}.csv', sep=';', encoding='ISO-8859-1', decimal=',')
        demo_dfp = pd.read_csv(f'{caminho_salvar_arquivos}/dfp_cia_aberta_{nome_demo}_con_{ano}.csv', sep=';', encoding='ISO-8859-1', decimal=',')

        #Remover o penúltimo registro
        demo_itr = demo_itr[demo_itr['ORDEM_EXERC'] == 'ÚLTIMO']
        demo_dfp = demo_dfp[demo_dfp['ORDEM_EXERC'] == 'ÚLTIMO']

        #Conversão dos tipos de dados
        demo_itr['VL_CONTA'] = pd.to_numeric(demo_itr['VL_CONTA'])
        demo_dfp['VL_CONTA'] = pd.to_numeric(demo_dfp['VL_CONTA'])

        demonstrativo = pd.concat([demonstrativo, demo_itr])
        demonstrativo = pd.concat([demonstrativo, demo_dfp])
        
    demonstrativo.to_csv(f'{caminho_salvar_arquivos}/{nome_demo}.csv', sep=';', encoding='ISO-8859-1', decimal=',', index=False)
    print(f'O demonstrativo {nome_demo} foi salvo com sucesso!')
```

Remover arquivos das demonstrações anuais

```python
list_arquivos = []
for diretorio, subpastas, arquivos in os.walk(caminho_salvar_arquivos):
    list_arquivos = arquivos
for de in demonstrativos:
    list_arquivos.remove(de+'.csv')

for arquivo in list_arquivos:
    try:
        os.remove(caminho_salvar_arquivos+'/'+arquivo)
        print(f'O {arquivo} foi removido com sucesso')
    except:
        print('Arquivo não encontrado')
```
