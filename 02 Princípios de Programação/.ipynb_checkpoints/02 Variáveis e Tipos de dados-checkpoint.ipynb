## Princípios de Programação e po Python

### Programa

Conjunto de instruções que contém as operações necessárias para, a partir de dados inseridos, obter um resultado que será disponibilizado por algum dispositivo de saída (XAVIER, 2018).

De modo geral, os programas de computadores são executados a partir da recepção de informações externada (Entrada), da execução das informações, com as quais foram previamente defindas pelo programados (Processamento) e do retorno dos dados após a finalização das instruções (Saída)



#### Entrada
* Dados envidados por teclado, leitores ópticos, leitura de arquivos texto ou de imagens...

#### Processamento
* Conjunto de instruções executados por uma linguagem de programação que foram anteriorementes escritas por programados.

#### Saída
* Retorno visuais nos monitores dos computadores, impressões, arquivos gravados em disco...

### Linguagem de Programação

Conjunto de repsentações e/ou regras utilizadas para criar programas. Por meio delas você estabelece uma comunicação com o computador, fazendo com que ele compreenda e execute o que você determinar (XAVIER, 2018).



### Lógica de Programação

É o uso correto das leis do pensamento, da 'ordem da razão' e dos processos de raciocínio e simbolização formais na programação de computadores (FORBELLONE, 2005). Com essas técnicas, é possível produzir soluções para problemas que são previamente sistematizados e estruturados.

O objetivo principal do estudo da Lógica de Programação é a construção de algoritmos coerentes e válidos.

### Algorítmos

Um algoritmo pode ser definido como uma seqüência de passos que visam a atingir um obje tivo bern definido (FORBELLONE, 2005).



### Por que Python?

* #### Produtividade do desenvolvedor

Python aumenta a produtividade do desenvolvedor comparado a linguagens como C, C++ e Java. O código em pythonn é, geralmente, 1/3 a 1/5 do tamanho do código equivalente às linguagens destacadas.

* #### Curva de aprendizagem

Python é uma linguagem de sintaxe simples comparados à outras como C, C++, Java e R.

* #### Bibliotecas

Python possui uma infinidade de funcionalidades pré-compiladas e portáveis, conhecidas como bibliotecas padrão.

* #### Análise e Ciência da Dados

Python é amplamente utilizada para a análise e ciência de dados por possuir bibliotecas consolidadas com Numpy, Pandas, Scikit-learn e Statsmodels.


### O que é nessário para programa em Python?

#### Instalação do intepretador Python

Como Python é uma linguagem intepretada, e, como tal, é um tipo de programa que executa outros programas. Quando você escreve programas em Python, o interpretador lê seu programa e executa as instruções.

Portanto, para começarmos a escrever nossos próprios programas devemos instalar no interpretador que está disponível gratuitamente no site: https://www.python.org/. Porém, para facilitar a instalação do interpretador Python, bem como demais ferramentas que utilizamos no curso, sugere-se a instalação do Software de Gerenciamento de Pacotes Anaconda. site: https://www.anaconda.com/products/individual



Após o download, deve-se seguir os passos da instalação semelhate a grande maioria de instalações que realizamos no dia a dia.



Com isso, temos disponível um software que facilita a instalação de outras ferramentas que o Jupyte notebook.

#### Jupyter notebook

O Projeto Jupyter é uma organização sem fins lucrativos criada para "desenvolver software de código aberto, padrões abertos e serviços para computação interativa em dezenas de linguagens de programação". Originado do IPython em 2014, o Projeto Jupyter suporta ambientes de execução em dezenas de linguagens de programação. O nome do projeto é uma referência às três principais linguagens de programação suportadas por Jupyter, Julia, Python e R.
